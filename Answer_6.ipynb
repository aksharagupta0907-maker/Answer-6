```python
import numpy as np

# Provided datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculate Covariance Matrix
# np.cov returns a matrix; [0, 1] gives the covariance between x and y
covariance = np.cov(list_x, list_y)[0, 1]

# Calculate Correlation Coefficient Matrix
# np.corrcoef returns a matrix; [0, 1] gives the correlation between x and y
correlation = np.corrcoef(list_x, list_y)[0, 1]

print(f"Covariance: {covariance}")
print(f"Correlation Coefficient: {correlation}")
```
